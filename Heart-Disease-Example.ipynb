```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file = './processed.cleveland.data.csv'
df = pd.read_csv(file)
columns = ['age','sex','cp','restbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','predicted_attribute']
df.columns = columns
```

## Explore the two data columns

```python
fig = plt.figure()

ax1 = fig.add_subplot(221)
ax1.set_title('Chest Pains')
ax1.hist(df['cp'])

ax2 = fig.add_subplot(222)
ax2.set_title("Predicted Attribute")
ax2.hist(df['predicted_attribute'])
```

## Plot the two hists over one another

```python
fig1, ax1 = plt.subplots()
ax1.set_title("Chest Pains & Predicted Attribute")
ax1.hist(df['cp'])
ax1.hist(df['predicted_attribute'])
```
